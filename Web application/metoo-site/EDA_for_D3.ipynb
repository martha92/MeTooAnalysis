```
import pandas as pd
import numpy as np

from google.colab import files
use_cuda = True
import numpy as np
from google.colab import drive

use_cuda = True
from google.colab import drive
drive.mount('/content/gdrive')

path = '/content/gdrive/My Drive/BigData_project/humans.csv'
df = pd.read_csv(path, delimiter=';',  encoding='utf8', engine='python', parse_dates=[3, 14])
```

```

df["date"]= df["created_date"].str.split(" ", n = 1, expand = True)[0] 
df['tweet_count'] = 1
df_filter = df[['id','date','name','followers_count','total_number_of_tweets','location','language','tweet_count']]
df_filter = df_filter.dropna(how='any')
df_filter
```

```
df_2017=df_filter[df_filter['date'].str.endswith('17')]

df1_2017 = df_2017[['name','followers_count','total_number_of_tweets','location','language']]
df1_2017.drop_duplicates(subset ="name",keep = 'first', inplace = True) 

groupby_tweetCount_df_2017 = df_2017.groupby(['name','location','language'],as_index=False)['tweet_count'].agg('sum')
groupby_tweetCount_df_2017 = groupby_tweetCount_df_2017.sort_values('tweet_count',ascending=False)

merged_df_2017 = pd.merge(left = df1_2017, right = groupby_tweetCount_df_2017, left_on='name', right_on ='name' , how='left')
merged_df_2017 = merged_df_2017.sort_values('tweet_count',ascending=False)

tweet_count_limit=1
result_df_2017=merged_df_2017[merged_df_2017['tweet_count'] > tweet_count_limit].reset_index(drop=True)

data_2017 = result_df_2017.drop(["location_x", "language_x"], axis=1)

data_2017.rename(columns={'location_y':'location'}, inplace=True)
data_2017.rename(columns={'language_y':'language'}, inplace=True)

file_name="users_2017.csv"
data_2017.to_csv(file_name, encoding='utf-8')
```
